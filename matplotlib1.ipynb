```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x = [1,2,3,4,5,6,9,32,23]
y = [2394,19247,19247,19247,19247,9127,934328,932874,23984]
```

```python
plt.plot(x,y,color="blue")
```

```python
plt.title("random graph")
plt.xlabel("random vals1")
plt.ylabel("random vals2")
plt.plot(x,y,color="blue")
```

** markersize is used to specify the marker thickness
** and alpha is used to define the transparency of the graph

```python
plt.title("random graph")
plt.xlabel("random vals1")
plt.ylabel("random vals2")
plt.plot(x,y,'gD-',markersize=3 , alpha=.1)
```

plotting many function in a graph 

```python
mint = [34,23,12,13,45,11]
maxt = [87,98,96,56,87,98]
days = [1,2,3,4,5,6]
avgt = [56,45,34,24,64,47]
```

```python
plt.plot(days,mint,'g-',label="min")
plt.plot(days,maxt,'r-',label="max")
plt.plot(days,avgt,'b-',label="avg")
plt.ylabel("temperature")
plt.xlabel("days")
plt.legend(loc="best" , fontsize="small" ,shadow="true")
plt.grid()
```
