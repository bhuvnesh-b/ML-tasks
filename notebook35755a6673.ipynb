```python
train, test = train_test_split(raw_data, test_size = 0.25)
print(train.shape)
print(test.shape)
```

```python
raw_test_data = pd.read_csv("../input/squid-beans/test.csv")
raw_test_data
```

```python
test_data = pd.get_dummies(raw_test_data) 
test_data = test_data.drop(columns = 'Bean ID')
test_data
```

```python
test_data.shape
```

```python
train_X = train[['MajorAxisLength','MinorAxisLength','ShapeFactor1',
                 'ShapeFactor2','ShapeFactor3','ShapeFactor4']]
train_Y = train.Class

test_X = test[['MajorAxisLength','MinorAxisLength','ShapeFactor1',
                 'ShapeFactor2','ShapeFactor3','ShapeFactor4']]


```

```python
train_X.head()
```

```python
test_Y.head()
```

```python
#Using GaussianNB
from sklearn.naive_bayes import GaussianNB
model3 = GaussianNB()
model3.fit(train_X,train_Y)
y_pred3 = model3.predict(test_X)

from sklearn.metrics import accuracy_score
print("Accuracy Score:",accuracy_score(test_Y,y_pred3))
```
