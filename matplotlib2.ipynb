```python
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np 
```

```python
company=['google' , 'amzn', 'msft' , 'fb']
revenue = [56,90,234,267]
```

```python
plt.bar(company,revenue,label="revenue")
```

```python
profit = [234,342,453,123]
```

```python
plt.bar(company,revenue,label="revenue")
plt.bar(company,profit,label="revenue")
plt.ylabel("revenue,profit")
plt.title("tech stocks")
```

```python
xp = np.arange(len(company))
xp
```

we can see the horizontal bar chrts also by writing barh instead of bar

```python
plt.bar(xp-.1,revenue,label="revenue",width=.2)
plt.bar(xp+.1,profit,label="profit",width=.2)
plt.xticks(xp,company)
plt.ylabel("revenue,profit")
plt.title("tech stocks")
plt.legend()
```
